# Make DataFrame

```python
import Predictor
import pandas as pd
import hvplot.pandas
import panel as pn
import sklearn
import sklearn.model_selection
import sklearn.metrics
import numpy as np
import xarray as xr
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

pn.extension()

Data = Predictor.Clean(filePath="C:\\Users\\sophi\\OneDrive\\Documents\\BigAssBudgetData.xlsx")
df = Data.copy()
df['DescCodes'], uniques = df['Description'].factorize()
df['CatCodes'], uniques = df['Category'].factorize()
```

# Make plot

```python
df.hvplot.scatter(x='Description', y='Amount', by='Category')
```

# Classify and Plot

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.inspection import DecisionBoundaryDisplay
data: pd.DataFrame = df[['DescCodes','Amount']]
target = df['Category']
X,y = data, target
X_Train, X_Test, y_train, y_test = sklearn.model_selection.train_test_split(X,y, test_size=0.3, random_state=42)
clf = make_pipeline(StandardScaler(),DecisionTreeClassifier(max_depth=5, random_state=42))
clf.fit(X_Train, y_train)
score = clf.score(X_Test, y_test)
```

```python

# plot the training Dataset
trainFrame = pd.DataFrame()
trainFrame[['Amount', 'DescCodes', ]] = X_Train[[ 'Amount', 'DescCodes', ]]
trainFrame['Category'] = y_train
trainFrame['Class'] = 'Train'
trainFrame.hvplot.scatter(x='DescCodes', y='Amount', by='Category', xlabel='Description Code')

```

```python

testFrame = X_Test[[  'DescCodes', 'Amount', ]]
testFrame['Category'] = y_test
testFrame['Class'] = 'Test'
testFrame.hvplot.scatter(x='DescCodes', y='Amount', by='Category')
```

```python
import matplotlib.pyplot as plt
from sklearn.svm import SVC
import holoviews
allData = pd.concat([trainFrame, testFrame])
allData.hvplot.scatter(x='DescCodes', y='Amount', by='Class')

model = SVC(kernel='rbf').fit(X_Train,y_train)

xMin, xMax = X_Train['DescCodes'].min() - 1, X_Train['DescCodes'].max() + 1
yMin, yMax = X_Train['Amount'].min() - 10, X_Train['Amount'].max() + 10
gridX, gridY = np.meshgrid(np.arange(xMin, xMax), np.arange(yMin, yMax))

gridPoints = np.c_[gridX.ravel(), gridY.ravel()]

Z = model.predict(gridPoints).reshape(gridX.shape)
GridFrame= pd.DataFrame({'gridX' : gridX.ravel(), 'gridY':gridY.ravel(), 'label': Z.ravel()}) 

boundary = GridFrame.hvplot.contour('gridX', 'gridY', z=Z) 
data = pd.DataFrame([X_Train, y_train])
data.hvplot.scatter()
(boundary * data).opts()

```
